```python
# Import Dependencies
import pandas as pd
import numpy as np
import datetime as dt
```

```python
# Read CSV Files & Create DataFrames
december_2017_df = pd.read_csv("data/201712-citibike-tripdata.csv")
december_2018_df = pd.read_csv("data/201812-citibike-tripdata.csv")
december_2019_df = pd.read_csv("data/201912-citibike-tripdata.csv")
```

```python
# Combined the 3 DataFrames Into a Single Combined DataFrame
december_combined_df = december_2017_df.append([december_2018_df, december_2019_df], ignore_index=True)
```

```python
# View Combined DataFrame
december_combined_df.head()
```

```python
# Rename Columns
december_combined_df = december_combined_df.rename(columns={"tripduration": "trip_duration",
                                                            "starttime": "start_time",
                                                            "stoptime": "stop_time",
                                                            "start station id": "start_station_id",
                                                            "start station name": "start_station_name",
                                                            "start station latitude": "start_station_latitude",
                                                            "start station longitude": "start_station_longitude",
                                                            "end station id": "end_station_id",
                                                            "end station name": "end_station_name",
                                                            "end station latitude": "end_station_latitude",
                                                            "end station longitude": "end_station_longitude",
                                                            "bikeid": "bike_id",
                                                            "usertype": "user_type",
                                                            "birth year": "birth_year"})
```

```python
# Convert Default Datetimes to Proper Datetime Format
december_combined_df["start_time"] = pd.to_datetime(december_combined_df["start_time"])
december_combined_df["stop_time"] = pd.to_datetime(december_combined_df["stop_time"])
```

```python
# View Combined DataFrame Renamed Columns & Reformatted Datetimes
december_combined_df.head()
```

```python
# Save Combined DataFrame to CSV
december_combined_df.to_csv("data/combined-citibike-tripdata.csv")
```
